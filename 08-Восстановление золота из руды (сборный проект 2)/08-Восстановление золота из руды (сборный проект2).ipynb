**Описание данных**
**Технологический процесс**
- Rougher feed — исходное сырье
- Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
- Xanthate **— ксантогенат (промотер, или активатор флотации);
- Sulphate — сульфат (на данном производстве сульфид натрия);
- Depressant — депрессант (силикат натрия).
- Rougher process (англ. «грубый процесс») — флотация
- Rougher tails — отвальные хвосты
- Float banks — флотационная установка
- Cleaner process — очистка
- Rougher Au — черновой концентрат золота
- Final Au — финальный концентрат золота

**Параметры этапов**
- air amount — объём воздуха
- fluid levels — уровень жидкости
- feed size — размер гранул сырья
- feed rate — скорость подачи

**Наименование признаков**
Наименование признаков должно быть такое:
[этап].[тип_параметра].[название_параметра]
Пример: rougher.input.feed_ag
Возможные значения для блока [этап]:
- rougher — флотация
- primary_cleaner — первичная очистка
- secondary_cleaner — вторичная очистка
- final — финальные характеристики
**Возможные значения для блока [тип_параметра]:**
- input — параметры сырья
- output — параметры продукта
- state — параметры, характеризующие текущее состояние этапа
- calculation — расчётные характеристики

### Проверка расчета эффективности обогащения

Для начала нам необходимо проверить правильно ли расчитана эффективность обогащения. Сделаем мы это на обучающей выборке для признака rougher.output.recovery и посчитаем MAE (среднюю абсолютную ошибку) между данными из таблицы и нашими расчетными данными.

Recovery = C * (F - T) / F * (C - T) * 100%

где:
- С - доля золота в концентрате после флотации/очистки;
- F - доля золота в сырье/концентрате до флотации/очистки;
- T - доля золота в отвальных хвостах после флотации/очистки.

```python
def recovery (row): # функция для расчета эффективности обогащения
    result = (row['rougher.output.concentrate_au'] *(row['rougher.input.feed_au']
                 - row['rougher.output.tail_au']))/(row['rougher.input.feed_au']
                 * (row['rougher.output.concentrate_au']-row['rougher.output.tail_au'])) *100
    
    return result
gold_recovery_train_new['roughet.calculation.recover'] = gold_recovery_train_new.apply(recovery, axis = 1)
```

Посчитаем среднюю абсолютную ошибку между исходными данными и рассчитанной нами величиной.

```python
mae = mean_absolute_error(gold_recovery_train_new['rougher.output.recovery'],
                          gold_recovery_train_new['roughet.calculation.recover'])

print("MAE:", mae)
```

Наша средняя обсюлютная ошибка достаточно мала, поэтому считаем, что данные по эффективности обогащения рассчитаны верно.

## Модель

### sMAPE

В качестве метрики качества для оценки наших моделей будем использовать sMAPE (симметричное среднее процентное отклонение). Для этого напишем функцию.


```python
def smape(target, predicted):
    
    return np.sum(abs(target - predicted) * 2 / ((abs(target) + abs(predicted)))) * 100/len(target)
```

В нашем анализе нам нужно предсказать два целевых признака, поэтому у нас появится такая метрика как "итоговое sMAPE". Для него тоже напишем функцию. 

```python
def smape_final(smape_r, smape_f):
        
    return 0.25 * smape_r + 0.75 * smape_f
```
